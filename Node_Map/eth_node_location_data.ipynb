```python
#Import packages and dependencies

import pandas as pd
from pathlib import Path
import folium
import geojson
from folium import plugins
import geopandas
from folium.plugins import MarkerCluster
```

```python
#Read in CSV file of node location data

eth_nodes = pd.read_csv(
    Path("../folium_eth_nodes_map/data/nodes-loc.csv"),
    index_col="host",
    skipinitialspace = False,
)

#Display the DataFrame
eth_nodes.head(30)
```

```python
#Split data from 'loc' column into new columns added to Dataframe: 'ip','success','address_type','continent', 'continent_code', 'country', 'country_code', 'country_flag', 'country_capital', 'country_phone', 'country_neighbors', 'region', 'city', 'latitude', 'longitude', 'asn', 'org', 'isp', 'timezone', 'timezone_name', 'timezone_dstoffset', 'timezone_gmtoffset', 'timezone_gmt', 'currency', 'currency_code', 'currency_symbol', 'currency_rates', 'currency_plural', 'completed_requests'

eth_nodes[['ip','success','address_type','continent', 'continent_code', 'country', 'country_code', 'country_flag', 'country_capital', 'country_phone', 'country_neighbors', 'region', 'city', 'latitude', 'longitude', 'asn', 'org', 'isp', 'timezone', 'timezone_name', 'timezone_dstoffset', 'timezone_gmtoffset', 'timezone_gmt', 'currency', 'currency_code', 'currency_symbol', 'currency_rates', 'currency_plural', 'completed_requests']] = eth_nodes['loc'].str.split(';', expand=True)

eth_nodes.head()

#Drop 'loc' column from Dataframe
eth_nodes_split = eth_nodes.drop('loc', axis=1)

eth_nodes_split
```

```python
#Drop timezone_gmt, timezone_dstoffset, timezone_gmtoffset, country_phone, and country_flag from Dataframe

eth_nodes_split.drop('timezone_gmt', axis=1, inplace=True)
eth_nodes_split.drop('timezone_dstoffset',axis=1, inplace=True)
eth_nodes_split.drop('timezone_gmtoffset',axis=1, inplace=True)
eth_nodes_split.drop('country_phone', axis=1, inplace=True)
eth_nodes_split.drop('country_flag', axis=1, inplace=True)

eth_nodes_split
```

```python
#Removing special characters and repeats of column headers in data

eth_nodes_split['ip'] = eth_nodes_split['ip'].str.replace("{'ip': '","")
eth_nodes_split['success'] = eth_nodes_split["success"].str.replace("'success': ","")
eth_nodes_split['address_type'] = eth_nodes_split["address_type"].str.replace("'type': ","")
eth_nodes_split['continent'] = eth_nodes_split["continent"].str.replace("'continent': ","")
eth_nodes_split['continent_code'] = eth_nodes_split["continent_code"].str.replace("'continent_code':","")
eth_nodes_split['country'] = eth_nodes_split["country"].str.replace("'country': ","")
eth_nodes_split['country_code'] = eth_nodes_split["country_code"].str.replace("'country_code': ","")
eth_nodes_split['country_code'] = eth_nodes_split["country_code"].str.replace("'country_code': ","")
eth_nodes_split['country_capital'] = eth_nodes_split["country_capital"].str.replace("'country_capital': ","")
eth_nodes_split['country_neighbors'] = eth_nodes_split["country_neighbors"].str.replace("'country_neighbors':","")
eth_nodes_split['country_neighbors'] = eth_nodes_split["country_neighbors"].str.replace(",","; ")
eth_nodes_split['region'] = eth_nodes_split["region"].str.replace("'region': ","")
eth_nodes_split['city'] = eth_nodes_split["city"].str.replace("'city': ","")
eth_nodes_split['latitude'] = eth_nodes_split["latitude"].str.replace("'latitude': ","")
eth_nodes_split['longitude'] = eth_nodes_split["longitude"].str.replace("'longitude': ","")
eth_nodes_split['asn'] = eth_nodes_split["asn"].str.replace("'asn': ","")
eth_nodes_split['org'] = eth_nodes_split["org"].str.replace("'org': ","")
eth_nodes_split['isp'] = eth_nodes_split["isp"].str.replace("'isp': ","")
eth_nodes_split['timezone'] = eth_nodes_split["timezone"].str.replace("'timezone': ","")
eth_nodes_split['timezone_name'] = eth_nodes_split["timezone_name"].str.replace("'timezone_name': ","")
eth_nodes_split['currency'] = eth_nodes_split["currency"].str.replace("'currency':","")
eth_nodes_split['currency_code'] = eth_nodes_split["currency_code"].str.replace("'currency_code': ","")
eth_nodes_split['currency_symbol'] = eth_nodes_split["currency_symbol"].str.replace("'currency_symbol': ","")
eth_nodes_split['currency_rates'] = eth_nodes_split["currency_rates"].str.replace("'currency_rates': ","")
eth_nodes_split['currency_plural'] = eth_nodes_split["currency_plural"].str.replace("'currency_plural': ","")
eth_nodes_split['completed_requests'] = eth_nodes_split["completed_requests"].str.replace("'completed_requests': ","")
eth_nodes_split['completed_requests'] = eth_nodes_split["completed_requests"].str.replace("}","")

eth_nodes_split = eth_nodes_split.replace("'","", regex=True)

eth_nodes_split.head(30)
```

```python
#Create DataFrame for Nodes located in United States

us_eth_nodes = eth_nodes_split[eth_nodes_split['country_code'].str.contains('US')]

us_eth_nodes
```

```python
# Generate heat map for US ETH Node locations
us_eth_node_heatmap = folium.Map(location = [35.227085,-80.843124], tiles='Cartodb dark_matter', zoom_start = 6)

# Create point geometries
geometry = geopandas.points_from_xy(us_eth_nodes.longitude, us_eth_nodes.latitude)
geo_df = geopandas.GeoDataFrame(us_eth_nodes[['ip','success','address_type','continent', 'country', 'country_code', 'country_capital', 'country_neighbors', 'region', 'city', 'latitude', 'longitude', 'asn', 'org', 'isp', 'timezone', 'completed_requests']], geometry=geometry)

heat_data = [[point.xy[1][0], point.xy[0][0]] for point in geo_df.geometry ]

heat_data
plugins.HeatMap(heat_data).add_to(us_eth_node_heatmap)

us_eth_node_heatmap
```

```python
# Generate map for marked US ETH Node locations
us_eth_node_locs = folium.Map(location = [35.227085,-80.843124], tiles='Cartodb dark_matter', zoom_start = 6)

# Create point geometries
geo_df_list = [[point.xy[1][0], point.xy[0][0]] for point in geo_df.geometry]

# Iterate through list and add a marker for each ETH Node IP Address.
i = 0
for coordinates in geo_df_list:
    type_color = "blue"

# Place the markers with the popup labels and data
    us_eth_node_locs.add_child(folium.Marker(location = coordinates,
                        popup =
                        "IP: " + str(geo_df.ip[i]) + '<br>' +
                        "Address Type: " + str(geo_df.address_type[i]) + '<br>' +
                        "Region: " + str(geo_df.region[i]) + '<br>'
                        "City: " + str(geo_df.city[i]) + '<br>'
                        "Org: " + str(geo_df.org[i]) + '<br>'
                        "ISP: " + str(geo_df.isp[i]) + '<br>'
                        "Completed Requests: " + str(geo_df.completed_requests[i]) + '<br>'
                        "Coordinates: " + str(geo_df_list[i]),
                        icon = folium.Icon(color = "%s" % type_color)))
    i = i + 1

us_eth_node_locs
```
