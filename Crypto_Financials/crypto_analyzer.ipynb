```python
#Import required libraries and dependencies

from pathlib import Path
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
# Read in Bitcoin csv as a Pandas Dataframe
btc_df = pd.read_csv(
    Path("Resources/BTC-USD.csv"),
    index_col="Date",
    parse_dates=True,
    infer_datetime_format=True
)
# Call the new Dataframe to insure the data was read in properly       
btc_df
```

```python
# Read in Ethereum CSV as a Pandas Dataframe
eth_df = pd.read_csv(
    Path("Resources/ETH-USD.csv"),
    index_col="Date",
    parse_dates=True,
    infer_datetime_format=True
)
# Call the new Dataframe to insure the data was read in properly
eth_df
```

```python
# Read in the S&P 500 csv as a Pandas Dataframe
sp_500_df= pd.read_csv(
    Path("Resources/SP_500_Historical_Data_1 (1).csv"),
    index_col="Date",
    parse_dates=True,
    infer_datetime_format=True
)
# Call the new Dataframe to insure the data was loaded in properly
sp_500_df
```

```python
# We dropped the 'Change %' in hopes to convert the data to 'float'
sp_500_df= sp_500_df.drop('Change %', axis=1)

sp_500_df
```

```python
# We dropped the "Vol." column in hopes it would let us convert the rest of the data to 'float'
sp_500_df=sp_500_df.drop('Vol.', axis=1)

sp_500_df
```

```python
# Used the 'dtypes' function to see the data types within the S&P dataframe
sp_500_df.dtypes
```

```python
# Use 'replace' function to drop all , from the S&P dataframe
sp_500_df = sp_500_df.replace(",","", regex=True)

sp_500_df
```

```python
#Use 'astype' to convert Objects in S&P dataframe to Float
sp_500_df = sp_500_df.astype('float')

sp_500_df.dtypes
```

```python
# Calculate percentage of change in ETH closing price
eth_daily_returns= eth_df.pct_change().dropna()

eth_daily_returns.head()
```

```python
# Calculate percentage of change in BTC closing price
btc_daily_returns= btc_df.pct_change().dropna()

btc_daily_returns.head()
```

```python
btc_volume = btc_df.loc[:, "Volume"]

btc_volume.head()
```

```python
eth_volume = eth_df.loc[:, "Volume"]

eth_volume.head()
```

```python
eth_volume.plot(legend=True, figsize=(15,10), title= "Bitcoin and Ethereum Market Cap 2021-22", color='orange', label='Ethereum')
btc_volume.plot(legend=True, figsize=(15,10), color='blue', label='Bitcoin')
```

```python
btc_df["Volatility"]=btc_df["Close"].pct_change().rolling(window=4).std().dropna()
btc_df.tail()
```

```python
eth_df["Volatility"]=eth_df["Close"].pct_change().rolling(window=4).std()
eth_df.tail()
```

```python
sp_500_df["Volatility"]=sp_500_df["Price"].pct_change().rolling(window=4).std().dropna()
sp_500_df.tail()
```

```python
sp500_volatility = sp_500_df.loc[:, "Volatility"]

sp500_volatility.tail()
```

```python
btc_volatility= btc_df.loc[:,"Volatility"]

btc_volatility.tail()
```

```python
eth_volatility=eth_df.loc[:, "Volatility"]

eth_volatility.tail()
```
